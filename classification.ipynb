```python
#before start upload to data/rss  all .csv files 

import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import plot_confusion_matrix, accuracy_score
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.metrics import plot_confusion_matrix, accuracy_score
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

pd.options.mode.chained_assignment = None
```

```python
# trim date:
#  @param: row String date for trimming
#  @param: dtype String  type of trimming: year: 2021, day: 01-01-2021
#  @return: date String
def cleanDate(row, dtype='year'):
    if (dtype == 'year'):
        return row.str.extract(r'(\d*)-\d*-\d*.*', expand=False)
    elif (dtype == 'day'):
        return row.str.extract(r'(\d*-\d*-\d*).*', expand=False)

# Change t and f to boolean:
#  @param: row String param for changing
#  @return: String boolean value    
def changeBoolean(row):
    d = {'t': True, 'f': False}
    return row.map(d)

# Fill NA total score row:
#  @param: row Object row of dataFrame for change
#  @return: row Object row of dataFrame with max studentScore from task
def fillNaTotalScore(row, fullDb):
    row['totalScore'] = fullDb[fullDb['courseTaskId'] == row['courseTaskId']]['studentScore'].max()
    return row

# Calculate sum points from task by studentId:
#  @param: studentId Integer student id for filtering
#  @param: task DataFrame with students tasks
#  @param: column String column for summing
#  return sum List points from task by studentId
def fillSum(studentId, task, column):
    return task[task['studentId'] == studentId][column].sum()


# Prepare englishLevel collumn, trim duplicates and clean data:
#  @param: clean_train DataFrame for change
#  @return: clean_train DataFrame with cleaned englishLevel column
def cleanLang(clean_train):
    clean_train['englishLevel'] = clean_train['englishLevel'].str.lower()
    clean_train['englishLevel'].replace(to_replace=r'^з$', value='-', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^natalya.sadov@gmail.com$', value='-', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^в1$', value='b1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^3/5$', value='b1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^good$', value='b2', regex=True, inplace=True)
    

    clean_train['englishLevel'].replace(to_replace=r'^.*begin.*$', value='a0', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*elementary.*$', value='a1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^базовый$', value='a0', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*начинающий.*$', value='a0', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^zero$', value='a0', regex=True, inplace=True)

    clean_train['englishLevel'].replace(to_replace=r'^easy$', value='a1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*на уровне чтения.*$', value='a1', regex=True, inplace=True)
    
    clean_train['englishLevel'].replace(to_replace=r'^.*pre-intermed.*$', value='a2', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*pre intermed.*$', value='a2', regex=True, inplace=True)
        
    clean_train['englishLevel'].replace(to_replace=r'^.*upper intermediate.*$', value='b2', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*upper-intermed.*$', value='b2', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*up-intermed.*$', value='b2', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*high-intermed.*$', value='b2', regex=True, inplace=True) 
    
    clean_train['englishLevel'].replace(to_replace=r'^.*interm.*$', value='b1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^средний$', value='b1', regex=True, inplace=True)  

    clean_train['englishLevel'].replace(to_replace=r'^.*advance.*$', value='c1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*proficient.*$', value='c1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*advanced.*$', value='c1', regex=True, inplace=True)    
    clean_train['englishLevel'].replace(to_replace=r'^.*high proficiency.*$', value='c1', regex=True, inplace=True)
    
    clean_train['englishLevel'].replace(to_replace=r'^native$', value='c2', regex=True, inplace=True)
    
 
    clean_train['englishLevel'].replace(to_replace=r'^b1-b2$', value='b1+', regex=True, inplace=True)    
    clean_train['englishLevel'].replace(to_replace=r'^a2-b1$', value='a2+', regex=True, inplace=True) 
    clean_train['englishLevel'].replace(to_replace=r'^b2-с1$', value='b2+', regex=True, inplace=True)    
    clean_train['englishLevel'].replace(to_replace=r'^a1-a2$', value='a1+', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^b2-c1$', value='b2+', regex=True, inplace=True)
    
    clean_train['englishLevel'].replace(to_replace=r'^.*a2\+.*$', value='a2+', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*a1\+.*$', value='a1+', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*а2\+.*$', value='a2+', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*с1.*$', value='a2+', regex=True, inplace=True)
    
    clean_train['englishLevel'].replace(to_replace=r'^a-2$', value='a2', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^а2$', value='a2', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^а1$', value='a1', regex=True, inplace=True)
    
    clean_train['englishLevel'].replace(to_replace=r'^a\+$', value='a2', regex=True, inplace=True)

    clean_train['englishLevel'].replace(to_replace=r'^.*b1[^\+]*$', value='b1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*b2[^\+]*$', value='b2', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*c1[^\+]*$', value='c1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*a1[^\+]*$', value='a1', regex=True, inplace=True)
    clean_train['englishLevel'].replace(to_replace=r'^.*a2[^\+]*$', value='a2', regex=True, inplace=True)

    return clean_train

# Prepare alias collumn, trim duplicates and clean data:
#  @param: df DataFrame for change
#  @return: df DataFrame with cleaned alias column
def cleanAlias(df):
    df['alias'].replace(to_replace=r'^.*react.*$', value='react', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*angular.*$', value='angular', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*android.*$', value='android', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*ios.*$', value='ios', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*nodejs.*$', value='nodejs', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*rs-st.*$', value='rss_st', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*rs-.*$', value='rss', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*js-fundamentals.*$', value='rss', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*st-.*$', value='rss_st', regex=True, inplace=True)
    df['alias'].replace(to_replace=r'^.*rss-pl.*$', value='rss_st', regex=True, inplace=True)
    return df
```

```python
# Prepare course dataframe: 
#  - Select COLUMNS_COURSE columns
#  - Change to boolean completed and inviteOnly columns
#  - Clean alias column
#      @param: df DataFrame for change
#      @return: df DataFrame with prepared data
def prepareCourse(df):
    COLUMNS_COURSE = [
        'courseId',
        'inviteOnly',
        'completed',
        'alias',
    ]
    
    df['completed'] = changeBoolean(df['completed'])
    df['inviteOnly'] = changeBoolean(df['inviteOnly'])
    df = cleanAlias(df)

    return df[df.columns.intersection(COLUMNS_COURSE)]
    

# Prepare task_results dataframe: 
#  - Select COLUMNS_TASK_RESULTS columns
#      @param: df DataFrame for change
#      @return: df DataFrame with prepared data
def prepareTaskResults(df):
    COLUMNS_TASK_RESULTS = [
        'courseTaskId',
        'studentId',
        'score',
    ]

    return df[df.columns.intersection(COLUMNS_TASK_RESULTS)]


# Prepare task dataframe: 
#  - Select COLUMNS_TASK columns
#  - Merge task and task_results. Give us new column score - user score for task. If score < 0 it 0 points
#  - Calculate studentScore and totalScore. 
#     If totalScore NA we found max student score and apply it value as totalScore per task
#     Calculate mean studentScore for same task.(crosscheck type).
#  - Drop duplicate tasks by courseTaskId and studentId. It drop crosscheck type of tasks, where different students put score for same task 
#  - Fix if studentScore > totalScore apply studentScore = totalScore 

#
#      @param: df DataFrame with tasks for change
#      @param: task_results DataFrame with task_results data.                           
#      @return: df DataFrame with prepared data
def prepareTask(df, task_results):

    COLUMNS_TASK = [
        'courseTaskId',
        'courseId',
        'scoreWeight',
        'maxScore',
        'totalScore',
    ]


    df = df[df.columns.intersection(COLUMNS_TASK)]

    df = pd.merge(task_results, df, left_on='courseTaskId', right_on='courseTaskId', how='left')


    df['score'] = pd.to_numeric(df['score'])
    df[df['score']<0] = 0
    
    df['studentScore'] = df['score']*df['scoreWeight']
    
    df['totalScore'] =  df.apply(lambda x: x['maxScore'] if x['courseId'] == 19 else x['maxScore']*x['scoreWeight'], axis = 1)

    df[df['totalScore'].isna()] = df[df['totalScore'].isna()].apply( lambda x: fillNaTotalScore(x, df), axis = 1)
    df['studentScore'] =  df.groupby(['studentId', 'courseTaskId'])['studentScore'].transform('mean')
    df = df.drop_duplicates(["studentId", "courseTaskId"]) 
    
    df['studentScore'] = df.apply(lambda x:  x['totalScore'] if x['studentScore']>x['totalScore'] else x['studentScore'], axis = 1)

    return df


# Prepare train and test dataframe: 
#  - Select COLUMNS_TRAIN columns
#  - Clean createdDate column. Stay only year
#  - Change isExpelled to boolean
#  - Calculate totalCurses and failCurses for each students.(select by userId)
#  - Calculate studentPoints, totalPoints and persPoints by course persPoints = studentPoints/totalPoints
#  - Calculate studentPoints totalPoints by course and persPoints studentPoints/totalPoints
#  - Fill NA mentorId and englishLevel
#  - Converting englishLevel to single format 
#  - Converting categorial features to str

#      @param: df DataFrame with train and test data
#      @param: course DataFrame with courses data.    
#      @param: task DataFrame with tasks data.  
#      @return: df DataFrame with prepared data
def prepareTrain(df, course, task):

    COLUMNS_TRAIN = [
        'isTopPerformer',
        'studentId',
        'createdDate',
        'isExpelled',
        'courseId',
        'userId',
        'mentorId',
        'englishLevel',
    ]
    df = df[df.columns.intersection(COLUMNS_TRAIN)]

    df['createdDate'] = cleanDate(df['createdDate'])
    df = pd.merge(df, course, left_on='courseId', right_on='courseId', how='left')

    df['isExpelled'] = changeBoolean(df['isExpelled'])


    df['totalCurses'] = df.groupby(['userId'])['userId'].transform('count')
    df['failCurses'] = df[df['isExpelled'] == False].groupby(['userId'])['userId'].transform('count')
    df['doneCurses'] = df[df['isExpelled'] == True].groupby(['userId'])['userId'].transform('count')

    df['failCurses'] = df['failCurses'].fillna(0)
    df['doneCurses'] = df['doneCurses'].fillna(0)

    df['failCursesTmp'] = df[df['failCurses'] == 0]['totalCurses'] - df[df['failCurses'] == 0]['doneCurses']
    df['failCursesTmp'] = df['failCursesTmp'].fillna(0)
    df['failCurses'] = df['failCurses']+df['failCursesTmp']
    df.drop(columns=['failCursesTmp', 'doneCurses'], inplace=True)


    df['englishLevel'] = df['englishLevel'].fillna('-')
    df['mentorId'] = df['mentorId'].fillna('-')

    df['studentPoints'] = df.apply(lambda x: fillSum(x['studentId'], task, 'studentScore') , axis = 1)
    df['totalPoints'] = df.apply(lambda x: fillSum(x['studentId'], task, 'totalScore') , axis = 1)
    df['totalPoints'] =  df.groupby(['courseId'])['totalPoints'].transform('max')
    
    df['persPoints'] = df['studentPoints']/df['totalPoints']
    
    categorical_features = [
        'createdDate',
        'isExpelled',
        'courseId',
        'userId',
        'mentorId',
        'englishLevel',
        'completed',
        'alias',
        'inviteOnly', 
    ]

    for column in categorical_features:
        df[column] = df[column].astype(str)   

    df = cleanLang(df)    
        
    return df
```

```python
PATH = "./data/"

course_original = pd.read_csv(os.path.join(PATH, 'rss', 'course.csv'))
task_original = pd.read_csv(os.path.join(PATH, 'rss', 'task.csv'))
task_results_original = pd.read_csv(os.path.join(PATH, 'rss', 'task_results.csv'))
train_original = pd.read_csv(os.path.join(PATH, 'rss', 'train.csv'))
test_original = pd.read_csv(os.path.join(PATH, 'rss', 'test.csv'))

task_results = prepareTaskResults(task_results_original)
task = prepareTask(task_original, task_results)
course = prepareCourse(course_original)

df_row_merged = pd.concat([train_original, test_original], ignore_index=True)
df_row_merged = prepareTrain(df_row_merged, course, task)

test = df_row_merged[df_row_merged['isTopPerformer'].isna()]
train = df_row_merged[df_row_merged['isTopPerformer'].isna() == False]
```

```python
CATEGORICAL_FEATURES =  [
        'createdDate',
        'isExpelled',
        'courseId',
        'userId',
        'mentorId',
        'englishLevel',
        'completed',
        'alias',
        'inviteOnly', 
    ]

NUMERIC_FEATURES  = [
        'totalCurses',
        'persPoints',         #top feature
        'failCurses',
        'studentPoints',      #top feature
        'totalPoints',
    ]



COLUMN_FOR_LEARN = np.concatenate([NUMERIC_FEATURES, CATEGORICAL_FEATURES], axis=0)

clean_train = train[train.columns.intersection(COLUMN_FOR_LEARN)]

X_train, X_test, y_train, y_test = train_test_split(clean_train, train['isTopPerformer'].to_list(), test_size=0.2, random_state=2020)

    
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps=[('encoder', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, NUMERIC_FEATURES),
        ('cat', categorical_transformer, CATEGORICAL_FEATURES)
    ])
X_train
```

```python
criterion_val           = 'gini'
max_depth_val           = 641 
min_samples_split_val   = 2
min_samples_leaf_val    = 1
n_estimators_val        = 231


pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier',   RandomForestClassifier(  n_estimators=n_estimators_val, 
                                     criterion=criterion_val,
                                     max_depth = max_depth_val,
                                     min_samples_split = min_samples_split_val,
                                     min_samples_leaf = min_samples_leaf_val))
])

#clf = pipeline.fit(X_train, y_train)
clf = pipeline.fit(clean_train, train['isTopPerformer'].to_list())
y_pred=clf.predict(X_test)


print('Result on scaled data: {:f}'.format(cross_val_score(clf, X_test, y_test, scoring='f1', cv=5).mean()))
```

```python
def saveResultToCsv(idVal, predictResult, filename='submission.csv'):
    submission = pd.DataFrame({'studentId':idVal, 'isTopPerformer':predictResult})
    submission['studentId'] = submission['studentId'].astype(int)
    submission.to_csv(filename,index=False) 

clean_test = test[test.columns.intersection(COLUMN_FOR_LEARN)]

y_pred=clf.predict(clean_test)

saveResultToCsv(test['studentId'],y_pred, 'out_final_gini2.csv')
```
